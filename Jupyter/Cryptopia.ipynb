```python
ñimport pandas as pd
import requests
import csv
import json
```

```python
market_history = requests.get("https://www.cryptopia.co.nz/api/GetMarketHistory/game_btc/999999" )     # <class 'dict'>
market_history = pd.DataFrame(market_history.json())                                            # all_markets.json() is class 'dict'

print("\nAPI Return as DF of JSON")
print(market_history.head(5))

payload = pd.DataFrame(list(market_history.Data))  
payload.index = pd.to_datetime(payload.Timestamp*10**9) # MIKE

print("\nAPI Return DF with JSON unpacked")
print(payload.head(5))
print(payload.tail(5))
```

```python

with open("/Users/michaelnew/Dropbox/Trading/Data/test_reverse.csv", 'w') as myfile:
    wr = csv.writer(myfile)
    wr.writerows(reversed(payload))
```
